```python
import matplotlib.pyplot as plt

numero = 1000000

def collatz(n):
  count =0
  while n>1:
    if n%2 == 0: #Se for par
      n = n/2      
    else: #se √çmpar
      n = 3*n+1
    count+=1
  return(count)
```

```python
lst_collatz =[]

for n in range(2,numero):
  lst_collatz.append(collatz(n))

#print(lst_collatz)
```

```python
x = list(range(2,numero))
```

```python
plt.figure(figsize=(15,10))
# plt.scatter(x[0:10000000], lst_collatz[0:10000000],color='darkgreen',marker='.')
plt.scatter(x, lst_collatz,color='darkgreen',marker='.')

```
